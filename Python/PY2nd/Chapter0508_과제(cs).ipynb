[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


변수명 = 데이터

다른 언어에서는 해당 변수에 저장할 데이터 타입에 맞추어 변수 타입을 지정해야 하지만(대표적으로 C) 파이썬은 저장되는 데이터에 따라 자동으로 데이터 타입을 결정해 주기 때문에 간편하다.

파이썬에서 모든 변수들은 객체를 할당받아 그 객체의 주소값을 가리키는 방식으로 메모리를 동적으로 할당받고, 각 객체에 할단된 레퍼런스의 개수가 0이되면 할당된 메모리는 가비지 컬렉션 기능으로 해제된다.

반면 C 언어는 대표적인 절차지향(Procedure-Oriented Programming, POP) 기법을 활용하며 개발자가 메모리의 할당과 해제를 관리한다. 정적 메모리 할당의 경우 컴파일시 고정된 크기 / 타입의 메모리 공간을 할당받아 사용하며 소멸시 운영체제가 자동으로 메모리를 반납하게 되고, 동적 메모리 할당도 가능하지만 (런타임에 원하는 크기/타입의 메모리 공간을 만들고, 해당 공간을 가리키는 주소를 포인터 변수에 저장하여 사용) 이 경우 메모리 반납은 수동으로 하여야 한다.

C 언어는 포인터를 사용하여 메모리 주소를 직접 조작할 수 있으며, 로우레벨 언어라고 할 수 있다. 파이썬은 변수가 객체에 대한 레퍼런스를 갖고 있으므로 변수 값을 변경할 시 메모리 값을 직접 조작하는 것이 아닌 변수의 레퍼런스 값만 변화한다.

C 언어는 여러 변수에 동시에 값을 할당하기 위해서는 여러 줄의 코드가 필요하며, 파이썬은 여러 변수에 동시에 값을 할당하기 비교적 용이하다.

[문제2]

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

컴퓨터는 이진 부동 소수점 표현법을 사용하는데, 이때 많은 십진 표현들이 이진 표현으로 정확하게 표기되지 않기 때문에 오류가 발생한다. 또한 파이썬에서 부동소수점 표현 시 유한 정밀도 한계로 인해 오류가 누적되거나 반올림, 부동 소수점 비교 오류등이 생긴다.

따라서 실수형을 비교할 때는 Decimal 이나 fractions 같이 좀 더 높은 정밀도를 가진 모듈을 사용하는 것이 좋다.
