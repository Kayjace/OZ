[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

얕은 복사 (Shallow Copy)
얕은 복사는 객체의 최상위 레벨은 복사해서 새로운 객체로 생성하지만, 하위 객체들(예: 리스트나 딕셔너리의 요소들)은 원본 객체와 참조를 공유함. 즉 얕은 복사의 내부 객체들은 원본 객체와 동일한 참조를 유지하기 때문에, 얕은 복사본의 내부 객체들을 변경 시 원본에 영향을 줄 수 있음.

깊은 복사 (Deep Copy)
깊은 복사는 객체와 그 안에 포함된 모든 객체를 재귀적으로 복사함. 즉, 깊은 복사는 원본 객체와 관련된 모든 객체들까지 새로운 객체로 생성하므로, 복사된 객체는 원본 객체와 완전히 독립적임.
