- [ ]  Python의 set에 대해 설명해 주세요
- [ ]  Python의 list는 연결리스트(linked list) 인가요?
- [ ]  딕셔너리 자료형에 대해 설명해주세요


리스트와 튜플의 주된 차이점은 가변성임. 리스트는 mutable, 튜플은 immutable. 튜플의 요소는 생성된 이후엔 추가, 삭제, 변경 불가. (타입에러 발생)


파이썬의 set는, 중복된 값을 가질 수 없는 자료형.(중복집합을 제외한 형태의 일반적인 집합이라고 볼 수도 있겠다.) 고유한 요소만을 저장하며 순서가 없음. 세트의 요소는 인덱싱을 통해 접근 가능함(iterable 하다고 볼 수 있을듯)
세트는 {}를 이용해 생성하거나 set()함수를 사용할 수 있음. 이름에 맞게 집합 연산을 수행하는데 유리함.

연결 리스트는 각 요소가 다음 요소를 가리키는 방식으로 구현되는 자료구조로, 요소의 추가와 삭제가 자유롭지만, 특정 요소에 접근하려면 처음부터 순차적으로 탐색해야 하는 단점이 있는 형태임. 파이썬의 리스트는 연결리스트가 아님. 동적 배열(dynamic array)로 구현되어있음. 이는 미리 일정량의 메모리를 할당받아, 필요에 따라 크기를 조절하는 데이터 구조. 메모리 할당을 두배씩 늘려서 받기 때문에 메모리 효율성은 튜플보다 떨어짐.

딕셔너리 자료형은 키(key)와 값(value)의 쌍을 요소로 가지는 자료형. 해시 테이블로 구현되어 있어, 키를 통해 값을 빠르게 검색할 수 있다. 중괄호 {} 안에 키:값 형태로 요소를 작성해서 생성함.
딕셔너리 키는 변경 불가, 딕셔너리 내에서 고유해야 함. set의 요소와 유사한 점이 있음.
