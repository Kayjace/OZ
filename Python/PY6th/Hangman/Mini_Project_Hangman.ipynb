Î¨∏Ï†úÍ∞Ä Ïñ¥Î†§Ïö∞Ïã† Í≤ΩÏö∞ Íµ¨Í∏ÄÎßÅ ÌÜµÌï¥ Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï¥Î≥¥ÏÑ∏Ïöî.  

```python
# Í∏∞Î≥∏ ÎùºÏù¥Î∏åÎü¨Î¶¨Îùº ÌååÏù¥Ïç¨ ÏÑ§ÏπòÏãú ÏûêÎèôÏúºÎ°ú ÏÑ§ÏπòÎê©ÎãàÎã§!
# random Î™®ÎìàÏùÄ ÎûúÎç§Ìïú Ïà´ÏûêÎ•º ÏÉùÏÑ±Ìï† Îïå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
import random

# Í≤åÏûÑÏóê ÏÇ¨Ïö©Îê† Îã®Ïñ¥ Î™©Î°ù
words = ["apple", "banana", "orange", "grape", "lemon"]

# ÌñâÎß® Í∑∏Î¶º
hangman_pics= [
'''
      +
      |
      |
      |
      |
      |
      |
      |
      |
=========''',
'''
  +---+
      |
      |
      |
      |
      |
      |
      |
      |
=========''',
'''
  +---+
  |   |
      |
      |
      |
      |
      |
      |
      |
=========''',
'''
  +---+
  |   |
  |   |
      |
      |
      |
      |
      |
      |
=========''', 
'''
  +---+
  |   |
  |   |
  |   |
      |
      |
      |
      |
      |
=========''', 
'''
  +---+
  |   |
  |   |
  |   |
  O   |
      |
      |
      |
      |
=========''', 
'''
  +---+
  |   |
  |   |
  |   |
  O   |
 /    |
      |
      |
      |
=========''',
'''
  +---+
  |   |
  |   |
  |   |
  O   |
 /|   |
      |
      |
      |
=========''', 
'''
  +---+
  |   |
  |   |
  |   |
  O   |
 /|\  |
      |
      |
      |
=========''' 
'''
  +---+
  |   |
  |   |
  |   |
  O   |
 /|\  |
 /    |
      |
      |
=========''', 
'''
  +---+
  |   |
  |   |
  |   |
  O   |
 /|\  |
 / \  |
      |
      |
=========''']

class HangmanGame:
    def __init__(self):
        self.word = random.choice(words)  # ÎûúÎç§ÏúºÎ°ú Îã®Ïñ¥ ÏÑ†ÌÉù

        self.guesses = set()  # Ï∂îÏ∏°Îêú Í∏ÄÏûê Ï†ÄÏû•

        self.attempts = 10  # ÏãúÎèÑ ÌöüÏàò

    def display(self):
        # Îã®Ïñ¥Ïùò ÌòÑÏû¨ ÏÉÅÌÉú ÌëúÏãú
        result = ""

        # ÏúÑÏóêÏÑú ÏÑ§Ï†ïÌïú Îã®Ïñ¥Ïùò Í∏ÄÏûêÎì§ÏùÑ ÌïòÎÇòÏî© Í∞ÄÏ†∏ÏòµÎãàÎã§.
        for char in self.word:
            # ÎßåÏïΩ Ï∂îÏ∏°Îêú Í∏ÄÏûêÍ∞Ä ÏÑ§Ï†ïÌïú Îã®Ïñ¥Ïóê Ìè¨Ìï®ÎêòÏñ¥ ÏûàÎã§Î©¥
            if char in self.guesses:
                # Ï∂îÏ∏°Îêú Í∏ÄÏûêÎ•º ÌëúÏãúÌï©ÎãàÎã§.
                result += char
            else:
                # Ï∂îÏ∏°Îêú Í∏ÄÏûêÍ∞Ä ÏïÑÎãàÎùºÎ©¥ _Î•º ÌëúÏãúÌï©ÎãàÎã§.
                result += "_"

        # Îã®Ïñ¥Ïùò ÌòÑÏû¨ ÏÉÅÌÉúÎ•º Î∞òÌôòÌï©ÎãàÎã§.
        return result

    def play(self):
        self.attempts=10
        # Í≤åÏûÑÏùÑ ÏãúÏûëÌï©ÎãàÎã§.
        # ÏãúÎèÑ ÌöüÏàòÍ∞Ä 0Ïù¥ Îê† ÎïåÍπåÏßÄ Î∞òÎ≥µ
        while self.attempts > 0:

            # ÌòÑÏû¨ Îã®Ïñ¥ ÏÉÅÌÉú ÌëúÏãú
            print(self.display())

            # Ï∂îÏ∏° ÌñàÎçò Í∏ÄÏûê Î≥¥Ïó¨Ï£ºÍ≥† ÏûÖÎ†•Î∞õÍ∏∞
            print(f"Ïù¥ÎØ∏ Ï∂îÏ∏°Ìïú Í∏ÄÏûê : {self.guesses}")
            guess = input("Í∏ÄÏûêÎ•º Ï∂îÏ∏°Ìï¥Î≥¥ÏÑ∏Ïöî: ").lower()

            # Ï∂îÏ∏°Ìïú Í∏ÄÏûêÍ∞Ä Ïù¥ÎØ∏ Ï∂îÏ∏°Ìïú Í∏ÄÏûêÎùºÎ©¥
            if guess in self.guesses:
                print("Ïù¥ÎØ∏ Ï∂îÏ∏°Ìïú Í∏ÄÏûêÏûÖÎãàÎã§.")

            # ÏÉàÎ°úÏù¥ Ï∂îÏ∏°Ìïú Í∏ÄÏûêÍ∞Ä ÏÑ§Ï†ïÌïú Îã®Ïñ¥Î•º ÎßûÏ∑ÑÎã§Î©¥
            elif guess in self.word:
                # Í∏ÄÏûêÎ•º Ï∂îÏ∏°ÌñàÎçò Í∏ÄÏûê Î™©Î°ùÏóê Ï∂îÍ∞Ä
                self.guesses.add(guess)
                # Îã®Ïñ¥Î•º ÎßûÏ∂îÏÖ®Íµ∞Ïöî. Îçî ÌûòÎÇ¥Î¥êÏöî"ÎùºÍ≥† Ï∂úÎ†•
                print(f"Îã®Ïñ¥Î•º ÎßûÏ∂îÏÖ®Íµ∞Ïöî. Îçî ÌûòÎÇ¥Î¥êÏöî.")
                # ÏÑ§Ï†ïÌïú Îã®Ïñ¥Í∞Ä Ï∂îÏ∏°ÌñàÎçò Í∏ÄÏûêÎì§Ïóê Ìè¨Ìï®ÎêúÎã§Î©¥(Ï†ïÎãµÏùÑ Î™®Îëê ÎßûÏ∑ÑÎã§Î©¥)
                if set(self.word).issubset(self.guesses):
                    print(f"Ï∂ïÌïòÌï©ÎãàÎã§! Îã®Ïñ¥Î•º ÎßûÏ∂îÏÖ®ÏäµÎãàÎã§: {self.word}")
                    break
            # Ï∂îÏ∏°Ìïú Í∏ÄÏûêÍ∞Ä ÏÑ§Ï†ïÌïú Îã®Ïñ¥Î•º ÎßûÏ∂îÏßÄ Î™ªÌñàÎã§Î©¥
            else:
                self.attempts -= 1
                print(f"ÌãÄÎ†∏ÏäµÎãàÎã§. ÎÇ®ÏùÄ ÏãúÎèÑ ÌöüÏàò: {self.attempts}")
                if self.attempts >= 0:
                    print(hangman_pics[10 - self.attempts])
            if self.attempts == 0:
                print(f"Í≤åÏûÑ Ïò§Î≤Ñ. Ï†ïÎãµÏùÄ: {self.word}")


if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
